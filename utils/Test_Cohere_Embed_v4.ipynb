```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

texts_map = {text: np.array(embedding) for text, embedding in zip(texts, response.embeddings)}

sources_map = {source: texts_map[source] for source in sources}
targets_map = {target: texts_map[target] for target in targets}

# Find top 3 closest targets for each source
top_matches = {}
for source, source_embedding in sources_map.items():
    similarities = []
    for target, target_embedding in targets_map.items():
        similarity = cosine_similarity([source_embedding], [target_embedding])[0][0]
        similarities.append((target, similarity))

    # Sort targets by similarity (highest first) and keep top 3
    top_3 = sorted(similarities, key=lambda x: x[1], reverse=True)[:3]
    top_matches[source] = top_3

# Example: print results
for source, matches in top_matches.items():
    print(f"Source: {source}")
    for i, (target, sim) in enumerate(matches, 1):
        print(f"  Match {i}: {target} (Similarity: {sim:.4f})")
    print()
```
